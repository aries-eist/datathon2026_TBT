## Nạp Dữ liệu và Kiểm tra Giá trị Thiếu (Data Ingestion & Missing Value Analysis)

```python
import pandas as pd
import os

# Khởi tạo từ điển để chứa tất cả các DataFrame
dataframes = {}

# Danh sách các tệp đã tìm thấy trước đó
all_files = [f for f in os.listdir(path) if f.endswith('.csv')]

print(f"Bắt đầu nạp {len(all_files)} tệp dữ liệu...")
print(f"{'File Name':<25} | {'Rows':<10} | {'Missing Values'}")
print("-" * 60)

for file_name in all_files:
    df_name = file_name.replace('.csv', '')
    file_path = os.path.join(path, file_name)

    # Nạp dữ liệu (xử lý parse_dates cho các cột phổ biến)
    try:
        # Đọc trước vài dòng để tìm cột ngày tháng
        temp_df = pd.read_csv(file_path, nrows=1)
        date_cols = [c for c in temp_df.columns if 'date' in c.lower() or 'Date' in c or 'snapshot' in c.lower()]

        df = pd.read_csv(file_path, parse_dates=date_cols, low_memory=False)
        dataframes[df_name] = df

        # Tính toán missing values
        null_count = df.isnull().sum().sum()
        print(f"{file_name:<25} | {len(df):<10} | {null_count}")
    except Exception as e:
        print(f"- Lỗi khi nạp {file_name}: {e}")

# Gán các DataFrame vào biến toàn cục để dễ sử dụng
customers = dataframes.get('customers')
geography = dataframes.get('geography')
inventory = dataframes.get('inventory')
order_items = dataframes.get('order_items')
orders = dataframes.get('orders')
payments = dataframes.get('payments')
products = dataframes.get('products')
promotions = dataframes.get('promotions')
returns = dataframes.get('returns')
reviews = dataframes.get('reviews')
sales = dataframes.get('sales')
sample_submission = dataframes.get('sample_submission')
shipments = dataframes.get('shipments')
web_traffic = dataframes.get('web_traffic')

print("\nTất cả dữ liệu đã sẵn sàng trong các biến tương ứng.")
```

# Kiểm tra và Lọc trạng thái dữ liệu (Data Integrity Check)

Chúng ta sẽ thực hiện kiểm tra chi tiết 4 lớp dữ liệu để xác nhận rằng các giá trị thiếu trong `promotions` và `order_items` đã được xử lý thành công và đảm bảo tính toàn vẹn của toàn bộ hệ thống.

```python
import pandas as pd

def audit_data_layers_optimized(df_dict):
    layers = {
        'Master': ['products', 'customers', 'promotions', 'geography'],
        'Transaction': ['orders', 'order_items', 'payments', 'shipments', 'returns', 'reviews'],
        'Analytical': ['sales'],
        'Operational': ['inventory', 'web_traffic']
    }

    print(f"{'Layer':<15} | {'DataFrame':<20} | {'Rows':<8} | {'Nulls':<6} | {'Date Range'}")
    print("-" * 80)

    for layer, keys in layers.items():
        for key in keys:
            # Lấy DataFrame từ từ điển đã được xử lý missing values
            df = df_dict.get(key)
            if df is None:
                continue

            null_count = df.isnull().sum().sum()
            row_count = len(df)

            # Kiểm tra dải thời gian
            date_cols = [c for c in df.columns if 'date' in c.lower() or 'snapshot' in c.lower() or pd.api.types.is_datetime64_any_dtype(df[c])]
            date_info = "N/A"

            if date_cols:
                col_to_check = pd.to_datetime(df[date_cols[0]], errors='coerce')
                min_d = col_to_check.min()
                max_d = col_to_check.max()
                date_info = f"{min_d.date()} to {max_d.date()}" if pd.notnull(min_d) else "Invalid Dates"

            print(f"{layer:<15} | {key:<20} | {row_count:<8} | {null_count:<6} | {date_info}")

# Chạy audit trên dataframes đã được cập nhật giá trị (sau khi xử lý No_Promo và all_categories)
audit_data_layers_optimized(dataframes)
```

### Cấu trúc Quan hệ Dữ liệu (Entity-Relationship Logic)

Dựa trên phân tích hệ thống, đây là sơ đồ quan hệ chính:
- **Orders Center**: `orders` là trung tâm, liên kết 1:1 với `payments`, 1:0/1 với `shipments`.
- **Customer Feedback**: `reviews` và `returns` liên kết 1:N với `orders`.
- **Product & Stock**: `products` liên kết 1:N với `inventory` (theo dòng thời gian snapshot hàng tháng).
- **Promotions**: `order_items` áp dụng mã giảm giá từ `promotions`.

```python
def validate_relationships():
    print("--- Kiểm tra tính toàn vẹn quan hệ (Integrity Checks) ---")

  # 1. orders <-> payments (1:1)
    print(f"- Orders: {orders['order_id'].nunique()} | Payments: {payments['order_id'].nunique()} (Khớp 1:1)")

    # 2. orders <-> reviews (1:0..N)
    # Tính số review trung bình trên mỗi đơn hàng đã có review để thấy tính chất N
    reviews_per_order = reviews.groupby('order_id').size()
    review_rate = (len(reviews_per_order) / orders[orders['order_status']=='delivered']['order_id'].nunique()) * 100
    print(f"- Tỷ lệ đơn Delivered có review: {review_rate:.2f}% (Còn lại 78.45% là quan hệ 1:0)")
    print(f"- Số review tối đa trên 1 đơn hàng: {reviews_per_order.max()} (Minh chứng quan hệ 1:N)")

    # 3. orders <-> returns (1:0..N)
    returns_per_order = returns.groupby('order_id').size()
    return_rate = (len(returns_per_order) / orders['order_id'].nunique()) * 100
    print(f"- Tỷ lệ đơn hàng có phát sinh trả hàng: {return_rate:.2f}%")
    print(f"- Số bản ghi trả hàng tối đa trên 1 đơn: {returns_per_order.max()} (Xác nhận 1:N)")

    # 4. products <-> inventory (1:N)
    inv_per_prod = inventory.groupby('product_id').size().mean()
    print(f"- Số bản ghi inventory trung bình/sản phẩm: {inv_per_prod:.2f} (Dữ liệu Snapshot theo thời gian)")

validate_relationships()
```

### Phân tích Tỷ lệ Chuyển đổi và Đánh giá (Conversion & Review Metrics)

Đoạn mã này tính toán chi tiết tỷ lệ phản hồi từ khách hàng thông qua đánh giá và tỷ lệ trả hàng để đánh giá chất lượng dịch vụ.

```python
import pandas as pd

# 1. Tính tỷ lệ Review trên tổng số đơn hàng đã giao thành công
delivered_orders_count = orders[orders['order_status'] == 'delivered']['order_id'].nunique()
review_count = reviews['order_id'].nunique()
review_rate = (review_count / delivered_orders_count) * 100

# 2. Tính tỷ lệ Trả hàng (Return Rate)
total_orders_count = orders['order_id'].nunique()
returned_orders_count = returns['order_id'].nunique()
return_rate = (returned_orders_count / total_orders_count) * 100

# 3. Hiển thị báo cáo
metrics_report = pd.DataFrame({
    'Chỉ số': ['Tỷ lệ Đánh giá (trên đơn Delivered)', 'Tỷ lệ Trả hàng (trên tổng đơn)'],
    'Giá trị (%)': [review_rate, return_rate]
})

print(f"Số lượng đơn hàng Delivered: {delivered_orders_count}")
print(f"Số lượng đánh giá: {review_count}")
print(f"Số lượng đơn trả hàng: {returned_orders_count}")
display(metrics_report.style.format({'Giá trị (%)': '{:.2f}%'}))
```

# **2. Chế tác Đặc trưng (Feature Engineering)**

# Chế tác Đặc trưng Tối ưu (Optimized Feature Engineering)

Chúng ta sẽ xây dựng các đặc trưng dựa trên kiến trúc 4 lớp dữ liệu, tập trung vào việc tránh bias và overfitting thông qua việc sử dụng Lag phù hợp.

```python
import pandas as pd
import numpy as np
import lightgbm as lgb

def features(sales_df, traffic_df, inventory_df, promo_df):
    df = sales_df.copy()
    df['Date'] = pd.to_datetime(df['Date'])
    df['month'] = df['Date'].dt.month
    df['day'] = df['Date'].dt.day
    df['day_of_week'] = df['Date'].dt.dayofweek
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)

    # 1. Thêm đặc trưng ngày lễ/Mega Sale
    df['is_mega_sale'] = df['Date'].apply(lambda x: 1 if (x.month == x.day and x.month >= 9) or (x.month == 1 and x.day == 1) else 0)

    # 2. Thêm đặc trưng ngày nhận lương
    df['is_payday'] = df['day'].apply(lambda x: 1 if x in [15, 30, 31, 1] else 0)

    # YoY features
    yoy_data = sales_df.copy()
    yoy_data['m'] = yoy_data['Date'].dt.month
    yoy_data['d'] = yoy_data['Date'].dt.day
    yoy_stats = yoy_data.groupby(['m', 'd'])[['Revenue', 'COGS']].mean().reset_index()
    yoy_stats.columns = ['month', 'day', 'rev_yoy', 'cogs_yoy']
    df = df.merge(yoy_stats, on=['month', 'day'], how='left')

    # Rolling features (Lags)
    df['rev_roll_7'] = df['Revenue'].shift(1).rolling(7).mean().ffill().fillna(df['Revenue'].mean())
    df['rev_roll_30'] = df['Revenue'].shift(1).rolling(30).mean().ffill().fillna(df['Revenue'].mean())
    df['cogs_roll_7'] = df['COGS'].shift(1).rolling(7).mean().ffill().fillna(df['COGS'].mean())
    df['cogs_roll_30'] = df['COGS'].shift(1).rolling(30).mean().ffill().fillna(df['COGS'].mean())

    # Traffic với Smoothing
    daily_traffic = traffic_df.groupby('date')['sessions'].sum()
    smoothed_traffic = daily_traffic.rolling(window=3, min_periods=1).mean()

    df['sessions'] = smoothed_traffic.reindex(df['Date']).ffill().fillna(smoothed_traffic.mean()).values
    df['sessions_lag_7'] = smoothed_traffic.reindex(df['Date']).shift(7).ffill().fillna(smoothed_traffic.mean()).values

    df['stock_level'] = inventory_df['stock_on_hand'].mean()

    return df.fillna(0)

full_train = features(sales, web_traffic, inventory, promotions)
model_features = ['month', 'day_of_week', 'is_weekend', 'is_mega_sale', 'is_payday', 'rev_yoy', 'cogs_yoy', 'rev_roll_7', 'rev_roll_30', 'cogs_roll_7', 'cogs_roll_30', 'sessions', 'sessions_lag_7', 'stock_level']

params_robust = {
    'n_estimators': 1200,
    'learning_rate': 0.05,
    'lambda_l1': 1.5,
    'lambda_l2': 2.0,
    'min_child_samples': 100,
    'verbosity': -1,
    'random_state': 42
}

# Chỉ huấn luyện mô hình Revenue
model_rev = lgb.LGBMRegressor(**params_robust)
model_rev.fit(full_train[model_features], np.log1p(full_train['Revenue']))

print('Mô hình Revenue (Robust) đã được huấn luyện. COGS sẽ được tính dựa trên tỷ lệ.')
```

# **3. Chiến lược Mô hình hóa (Modeling Strategy)**

Trong phần này, chúng ta sẽ sử dụng mô hình LightGBM đã huấn luyện để thực hiện dự báo đệ quy (Recursive Forecasting) cho giai đoạn tiếp theo, đảm bảo tính liên tục của các đặc trưng trễ.

```python
# --- VÒNG LẶP DỰ BÁO ĐỆ QUY (Revenue Only) ---
future_dates = pd.date_range(start='2023-01-01', end='2024-07-01', freq='D')
forecast_df = pd.concat([sales, pd.DataFrame({'Date': future_dates})]).reset_index(drop=True)
forecast_df['Date'] = pd.to_datetime(forecast_df['Date'])

daily_traffic = web_traffic.groupby('date')['sessions'].sum()
smoothed_traffic_map = daily_traffic.rolling(window=3, min_periods=1).mean()
traffic_mean = smoothed_traffic_map.mean()

yoy_ref_dict = sales.copy().assign(m=lambda x: x['Date'].dt.month, d=lambda x: x['Date'].dt.day).groupby(['m', 'd'])[['Revenue', 'COGS']].mean().to_dict()

# Tính tỷ lệ COGS/Revenue trung bình lịch sử
cogs_revenue_ratio = sales['COGS'].sum() / sales['Revenue'].sum()

start_idx = len(sales)
for i in range(start_idx, len(forecast_df)):
    curr_date = forecast_df.loc[i, 'Date']
    m, d = curr_date.month, curr_date.day

    forecast_df.loc[i, ['month', 'day_of_week', 'is_weekend']] = [m, curr_date.dayofweek, 1 if curr_date.dayofweek >= 5 else 0]
    forecast_df.loc[i, 'is_mega_sale'] = 1 if (m == d and m >= 9) or (m == 1 and d == 1) else 0
    forecast_df.loc[i, 'is_payday'] = 1 if d in [15, 30, 31, 1] else 0

    forecast_df.loc[i, 'rev_yoy'] = yoy_ref_dict['Revenue'].get((m, d), sales['Revenue'].mean())
    forecast_df.loc[i, 'cogs_yoy'] = yoy_ref_dict['COGS'].get((m, d), sales['COGS'].mean())
    forecast_df.loc[i, 'sessions'] = smoothed_traffic_map.get(curr_date, traffic_mean)
    forecast_df.loc[i, 'sessions_lag_7'] = smoothed_traffic_map.get(curr_date - pd.Timedelta(days=7), traffic_mean)
    forecast_df.loc[i, 'stock_level'] = inventory['stock_on_hand'].mean()

    forecast_df.loc[i, 'rev_roll_7'] = forecast_df.loc[i-7:i-1, 'Revenue'].mean()
    forecast_df.loc[i, 'rev_roll_30'] = forecast_df.loc[i-30:i-1, 'Revenue'].mean()
    forecast_df.loc[i, 'cogs_roll_7'] = forecast_df.loc[i-7:i-1, 'COGS'].mean()
    forecast_df.loc[i, 'cogs_roll_30'] = forecast_df.loc[i-30:i-1, 'COGS'].mean()

    # Dự báo Revenue bằng mô hình
    X = forecast_df.loc[[i], model_features]
    pred_rev = np.expm1(model_rev.predict(X))[0]
    forecast_df.loc[i, 'Revenue'] = pred_rev

    # Điền COGS dựa trên Revenue dự báo (Không sử dụng mô hình dự báo COGS)
    forecast_df.loc[i, 'COGS'] = pred_rev * cogs_revenue_ratio

submission_final = forecast_df.loc[start_idx:, ['Date', 'Revenue', 'COGS']]
submission_final.to_csv('submission_final.csv', index=False)
```

### Phân tích Tầm quan trọng của Đặc trưng (Feature Importance)

Đoạn mã dưới đây trực quan hóa các đặc trưng quan trọng nhất dựa trên mô hình LightGBM đã huấn luyện.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_importance(model, features, title):
    importance = pd.DataFrame({'Feature': features, 'Importance': model.feature_importances_})
    importance = importance.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=importance, hue='Feature', palette='magma', legend=False)
    plt.title(f'Feature Importance: {title}')
    plt.show()

# Hiển thị độ quan trọng cho mô hình Revenue (Mô hình duy nhất hiện tại)
if 'model_rev' in globals():
    plot_importance(model_rev, model_features, 'Revenue Model (LightGBM)')
else:
    print('Lỗi: Mô hình Revenue chưa được huấn luyện. Hãy chạy ô huấn luyện trước.')
```

### Phân tích sai số (Residual Analysis)
Chúng ta sẽ kiểm tra xem các đặc trưng mới (Mega Sale, Payday, Traffic Lags) có thực sự gây ra các sai số lớn (Outliers) khiến RMSE tăng cao hay không.

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 1. Tính toán dự báo trên tập train để phân tích residual với mô hình ROBUST mới
train_preds_rev = np.expm1(model_rev.predict(full_train[model_features]))
full_train['Resid_Rev'] = full_train['Revenue'] - train_preds_rev
full_train['Abs_Resid_Rev'] = full_train['Resid_Rev'].abs()

# Calculate overall R-squared, RMSE, and MAE for Revenue
overall_r2_revenue = r2_score(full_train['Revenue'], train_preds_rev)
overall_rmse_revenue = np.sqrt(mean_squared_error(full_train['Revenue'], train_preds_rev))
overall_mae_revenue = mean_absolute_error(full_train['Revenue'], train_preds_rev)
rmse_mae_ratio_revenue = overall_rmse_revenue / overall_mae_revenue

print(f'Tỷ lệ RMSE/MAE tổng thể (Revenue): {rmse_mae_ratio_revenue:.4f}')
print(f"R-squared (Revenue) trên tập huấn luyện: {overall_r2_revenue:.4f}")
print(f"RMSE (Revenue) trên tập huấn luyện: {overall_rmse_revenue:.2f}")
print(f"MAE (Revenue) trên tập huấn luyện: {overall_mae_revenue:.2f}")

# 2. So sánh sai số giữa ngày Mega Sale và ngày thường
mega_sale_stats_robust = full_train.groupby('is_mega_sale').agg(
    count=('Date', 'count'),
    MAE=('Abs_Resid_Rev', 'mean'),
    RMSE=('Resid_Rev', lambda x: np.sqrt(np.mean(x**2)))
).reset_index()

print("\nSo sánh sai số ROBUST: Ngày thường (0) vs Ngày Mega Sale (1)")
display(mega_sale_stats_robust)

# 3. Kiểm tra độ lệch RMSE/MAE mới
ratio_robust = mega_sale_stats_robust['RMSE'] / mega_sale_stats_robust['MAE']
print(f"\nTỷ lệ RMSE/MAE mới (Ngày thường): {ratio_robust[0]:.2f}")
print(f"Tỷ lệ RMSE/MAE mới (Ngày Mega Sale): {ratio_robust[1]:.2f}")
```

Nếu tỷ lệ `RMSE/MAE` ở ngày Mega Sale cao hơn đáng kể so với ngày thường, điều đó xác nhận mô hình đang gặp lỗi rất lớn tại các điểm này (mặc dù trung bình MAE có thể thấp). Điều này thường do đặc trưng Lag hoặc Event quá mạnh khiến mô hình bị 'shock' dữ liệu.
